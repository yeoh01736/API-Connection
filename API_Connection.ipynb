```python
# Another API - Space Images

import requests

url = "https://images-api.nasa.gov/search?q=earth"

response = requests.get(url)
data = response.json()

# Show the first result title and image URL
first_item = data["collection"]["items"][0]
title = first_item["data"][0]["title"]
image_url = first_item["links"][0]["href"]

print("üõ∞Ô∏è Title:", title)
print("üñºÔ∏è Image URL:", image_url)
```

# Loops through the top 5 image results from NASA

```python
import requests

# Step 1: Search for something (e.g., "mars")
query = "galaxy"
url = f"https://images-api.nasa.gov/search?q={query}"

response = requests.get(url)
data = response.json()

# Step 2: Loop through the first 5 items
items = data["collection"]["items"]

for i in range(5):
    try:
        item = items[i]
        title = item["data"][0]["title"]
        image_url = item["links"][0]["href"]

        print(f"\nüöÄ Result {i+1}")
        print("üñºÔ∏è Title:", title)
        print("üîó Image URL:", image_url)
    except Exception as e:
        print(f"\n‚ö†Ô∏è Could not process result {i+1}: {e}")
```
